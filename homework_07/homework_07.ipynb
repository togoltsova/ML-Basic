### Импортируйте NumPy

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr = numpy.zeros(10, dtype=int)
arr[4] = 1
arr_2d = arr.reshape(1, -1)
print(arr)
print(arr_2d)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr_3 = numpy.arange(10, 50)
arr_reverse = arr_3[::-1]
arr_evens = arr_reverse[arr_reverse % 2 == 0]
print(arr_3)
print(arr_reverse)
print(arr_evens)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr_4 = numpy.arange(0, 9).reshape(3, 3)
print(arr_4)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr_5 = numpy.random.randint(0, 24, (4, 3, 2))
print(arr_5)
print(arr_5.min())
print(arr_5.max())
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr_6 = numpy.arange(0, 24).reshape(6, 4)
arr_7 = numpy.arange(0, 12).reshape(4, 3)
print(arr_6)
print(arr_7)
print(numpy.dot(arr_6, arr_7))
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr_8 = numpy.random.randint(0, 49, (7,7))
print(arr_8)
mean = arr_8.mean()
print(mean)
std = arr_8.std()
print(std)
arr_normalized = (arr_8 - mean) / std
print(arr_normalized)
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
tips_mean_total_bill_by_day = tips.groupby("day")["total_bill"].mean()
print(tips_mean_total_bill_by_day)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
tips_total_bill_median = tips["total_bill"].median()
print(tips_total_bill_median)
tips_mean_tip_by_day = tips[tips["total_bill"] > tips_total_bill_median].groupby("day")["tip"].mean()
print(tips_mean_tip_by_day)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips["smoker"] = (tips["smoker"] == "Yes").astype("int")
tips.head()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
tips = seaborn.load_dataset('tips')
tips_relplot = seaborn.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="sex",
    hue="smoker",
    palette={"No" : "tab:blue", "Yes": "tab:orange"},
    kind="scatter",
    height=4,
    aspect=1
)

tips_relplot.set_axis_labels("total_bill", "tip")
tips_relplot.set_titles("{col_name}")
matplotlib.pyplot.show()

```
