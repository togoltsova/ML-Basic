## 1. Даны матрицы:

$$
A =
\begin{pmatrix}
2 & -4 \\
3 & 5 \\
-1 & 0
\end{pmatrix},
\quad
B =
\begin{pmatrix}
1 & 2 & 7 \\
-3 & -4 & 0 \\
5 & 2 & 1
\end{pmatrix},
\quad
C =
\begin{pmatrix}
6 & -3 & 9 \\
4 & -5 & 2 \\
8 & 1 & 5
\end{pmatrix}
$$

Посчитайте матрицу

$$
D = A^T C - 2A^T B^T
$$

Приведите полную последовательность вычислений.


Решение:

1)
$$
A^T =
\begin{pmatrix}
2 & 3 & -1 \\
-4 & 5 & 0
\end{pmatrix}
$$

2)
$$
B^T =
\begin{pmatrix}
1 & -3 & 5 \\
2 & -4 & 2 \\
7 & 0 & 1
\end{pmatrix}
$$

3)
$$
A^T \times C =
\begin{pmatrix}
2 & 3 & -1 \\
-4 & 5 & 0
\end{pmatrix}
\times
\begin{pmatrix}
6 & -3 & 9 \\
4 & -5 & 2 \\
8 & 1 & 5
\end{pmatrix}
=
\begin{pmatrix}
2*6 + 3*4 + (-1)*8 & 2*(-3) + 3*(-5) + (-1)*1 & 2*9 + 3*2 + (-1)*5 \\
-4*6 + 5*4 + 0*8 & -4*(-3) + 5*(-5) + 0*1 & -4*9 + 5*2 + 0*5
\end{pmatrix}
=
\begin{pmatrix}
16 & -22 & 19 \\
-4 & -13 & -26
\end{pmatrix}
$$

4)
$$
A^T \times B^T =
\begin{pmatrix}
2 & 3 & -1 \\
-4 & 5 & 0
\end{pmatrix}
\times
\begin{pmatrix}
1 & -3 & 5 \\
2 & -4 & 2 \\
7 & 0 & 1
\end{pmatrix}
=
\begin{pmatrix}
2*1 + 3*2 + (-1)*7 & 2*(-3) + 3*(-4) + (-1)*0 && 2*5 + 3*2 + (-1)*1 \\
-4*1 + 5*2 + 0*7 & (-4)*(-3) + 5*(-4) + 0*0 && (-4)*5 + 5*2 + 0*1 \\
\end{pmatrix}
=
\begin{pmatrix}
1 & -18 & 15 \\
6 & -8 & -10
\end{pmatrix}
$$

5)
$$
D = A^T \times C - 2A^T \times B^T =
\begin{pmatrix}
16 & -22 & 19 \\
-4 & -13 & -26
\end{pmatrix}
- 2 \times
\begin{pmatrix}
1 & -18 & 15 \\

## 3.  Укажите те значения параметров $p$ и $q$, при которых ранг матрицы

$$
A =
\begin{pmatrix}
1 & 2 & 3 \\
3 & 6 & p \\
5 & 10 & q
\end{pmatrix}
$$

равен единице.

Решение:
ранг матрицы = это мах количество линейно независимых строк.
соответсвенно, для ранга = 1, нужно привести матрицу А к виду:
$$
A =
\begin{pmatrix}
a1 & a2 & a3 \\
0 & 0 & 0 \\
0 & 0 & 0
\end{pmatrix}
$$
1) Арифметические операции между строками не изменяют ранг матрицы. <br>
2) Домножим первую строку на 3 и вычтем из нее вторую строку. <br>
$$
A =
\begin{pmatrix}
1 & 2 & 3 \\
0 & 0 & 9-p \\
5 & 10 & q
\end{pmatrix}
$$
3) Домножим первую строку на 5 и вычтем из нее третью строку. <br>
$$
A =
\begin{pmatrix}
1 & 2 & 3 \\
0 & 0 & 9-p \\
0 & 0 & 15-q
\end{pmatrix}
$$
4)
9-p = 0 <br>
15-q = 0 <br>
5. <br>
p = 9 <br>
q = 15 <br>
