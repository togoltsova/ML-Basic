### 5. Исследовательское задание: малоранговая аппроксимация матрицы

Сгенерируйте случайную квадратную матрицу
$$
A \in R^{n \times n}, \quad n \ge 100.
$$

С помощью функции numpy.linalg.svd выполните сингулярное разложение этой матрицы и получите три матрицы:
$$
U, S, V^T.
$$

Выполняйте аппроксимацию матрицы $A$ с рангом $r$, меняя его значение, например, от $2$ до $n$:

$$
\widetilde{A}
=
U[:, :r]\, S[:r, :r]\, V^T[:r, :].
$$

и каждый раз вычисляйте ошибку аппроксимации (как восстановленная матрица отличается от исходной):

$$
E(r)
=
\|A - \widetilde{A}\|_F
=
\sqrt{
\sum_{i=1}^{n}
\sum_{j=1}^{n}
\left(a_{ij} - \widetilde{a}_{ij}\right)^2
}.
$$

Используя библиотеку matplotlib, постройте график зависимости ошибки аппроксимации матрицы от ранга $r$.

Решение:

```python
# 1) Генерируем матрицу
import numpy as np

np.random.seed(3)
n = 100
A = np.random.randn(n, n)
# Смотрим, что сгенерили
print(A.shape)
print(A[:3, :3])
```

```python
# 2) Сингулярное разложение матрицы А на три матрицы (U, S, Vт) помощью функции numpy.linalg.svd.
U, s, Vt = np.linalg.svd(A)
# смотрим, что получилось
print(U.shape)
print(s.shape)
print(Vt.shape)
```

```python
# видим, что s не матрица
# смотрим документацию, почему так
help(np.linalg.svd)
```

```python
# функция numpy.linalg.svd возвращает s : (..., K) array = Vector(s) with the singular values.
# значит, нужно преобразовать этот вектор в матрицу.
# в теории svd, S - диагональная матрица... соответсвенно, сделаем матрицу S с диагональю из значений вектора s
S = np.diag(s)
print(S.shape)
```

```python
# 3) Выполняем аппроксимацию матрицы A с рангом r, меняя его значение, от 2 до n
# Берём первые r:
# из U первые r столбцов U_r = U[:, :r]
# из S верхний левый r×r блок S_r = S[:r, :r]
# из Vt первые r строк Vt_r = Vt[:r, :]
# И перемножаем: A_r = U_r @ S_r @ Vt_r.
# посмоотрим для r = 2
U_r = U[:, :2]
S_r = S[:2, :2]
Vt_r = Vt[:2, :]
# смотрим, что получилось
print(U_r.shape)
print(S_r.shape)
print(Vt_r.shape)
```

```python
# Найдем апроксимацию A с рангом r, меняя его значение, от 2 до n
A_r_array = []
for r in range(2, n):
    A_r = U[:, :r] @ S[:r, :r] @ Vt[:r, :]
    A_r_array.append(A_r)
# Смотрим что получилось
print(A_r_array[:1])
```

```python
# чудесно. получили массив матриц
```

```python
# теперь посчитаем вместе с ошибкой апроксимации (как восстановленная матрица отличается от исходной) по норме Фробениуса.
# используем функцию np.linalg.norm(A - A_r, ord='fro')
errors = []
for r in range(2, n):
    A_r = U[:, :r] @ S[:r, :r] @ Vt[:r, :]
    err = np.linalg.norm(A - A_r, ord='fro')
    errors.append(err)
# Смотрим что получилось
print(errors[:5])
```

```python
# получили массив ошибок в числовом виде.
# 4) используя библиотеку matplotlib, построим график зависимости ошибки аппроксимации матрицы от ранга $r$.
import matplotlib.pyplot as plt
plt.figure()
plt.plot(list(range(2, n)), errors)
plt.xlabel("rank")
plt.ylabel("error")
plt.title("зависимость ошибки аппроксимации матрицы от ранга r")
plt.grid(True)
plt.show()
```

Видим график, похожий на гиперболу при положительном х.<br>
Что это значит?<br>
Это означает обратную зависимоть ошибки от ранга => чем больше ранг матрицы, тем меньше ошибка аппроксимации.<br>
Зачем мы это задание делали? svd = это про сжатие данных, сколько можно выкинуть, не ухудшив серьезно качество данных. любопытно, что в случае данной задачи, график показывает, что минимальной ошибке соответсвует ранг близкий к 100, а это полный размер матрицы. значит, если начать сжимать данные, ошибка быстро начинает расти. ### 2. Проверьте, что функция
$$
f(x_1, x_2) = \ln(\sqrt{x_1} + \sqrt{x_2})
$$

удовлетворяет уравнению

$$
x_1 \frac{\partial f}{\partial x_1} +
x_2 \frac{\partial f}{\partial x_2}
= \frac{1}{2}.
$$

Решение:<br>
Проиводная сложной функции равна произведению производной внешней функции на производную внутренней функции.<br>
Обозначим внутренню $(\sqrt{x_1} + \sqrt{x_2}) = g$, тогда внешняя функция имеет вид: <br>
$$
f(g) = \ln(g)
$$
Производная внешней функции равна 1/g
<br>
Найдем производную внутренней функции по $ \frac{\partial g}{\partial x_1}$, считая $x_2$ константой.
Проихводная $\sqrt{x_1} = 1/2 (\sqrt{x_1})^{1/2}$
$$
\frac{\partial g}{\partial x_1} = \frac{1}{2\sqrt{x_1}}
$$
Найдем производную внутренней функции по $ \frac{\partial g}{\partial x_2}$, считая $x_1$ константой.
Проихводная $\sqrt{x_2} = 1/2 (\sqrt{x_2})^{1/2}$
$$
\frac{\partial g}{\partial x_2} = \frac{1}{2\sqrt{x_2}}
$$
Вернемся к производной внешней функции (1/g), соответсвенно, проиводные сложной функции по $x_1$ и $x_2$: <br>
$$
\frac{\partial f}{\partial x_1} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} * \frac{1}{2\sqrt{x_1}} = \frac{1}{2\sqrt{x_1} * (\sqrt{x_1} + \sqrt{x_2})}
$$
$$
\frac{\partial f}{\partial x_2} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} * \frac{1}{2\sqrt{x_2}} = \frac{1}{2\sqrt{x_2} * (\sqrt{x_1} + \sqrt{x_2})}
$$
<br>

# 4. (Куб евклидовой нормы). Найдите первый дифференциал $df(x)$ и градиент $\nabla f(x)$ функции:

$$
f(x) = \frac{1}{3}\|x\|_2^3, \qquad x \in \mathbb{R}^n \setminus \{0\}.
$$

Решение: <br>
Пусть $u = <x,x> $, тогда $\|x\|_2 = \sqrt{u} $,
$$
f(u) = 1/3 \sqrt{u}^3
$$
$$
\frac{df}{du} = f′(u) = 1/3 * 3/2 * u^{1/2} = 1/2 \sqrt{u}
$$
$$
df = f′(u) du
$$
$$
df = 1/2 \sqrt{u} du
$$


Подставляем $u = <x,x> $, находим дифференциал $du$
$$
du = d<x,x> = <dx, x> + <x, dx>
$$
скалярное произведение симметрично: $⟨dx,x⟩ = ⟨x,dx⟩$, тогда
$$
du = d<x,x> = 2<x, dx>
$$
находим дифференциал $df(x)$
$$
df(x) = 1/2 \sqrt{<x,x>} * 2<x, dx> = 1/2 \|x\|_2 * 2<x,dx> = \|x\|_2 <x,dx>
$$
$<x,dx>$ скалярное произведение => можно представить как произведение строки $x^T$ на столбец $dx$
$$
df(x) = \|x\|_2 x^T dx
$$

Находим градиент $\nabla f(x)$ из $ df = <\nabla f(x), dx> = (\nabla f(x))^T dx$
$$
(\nabla f(x))^T = \|x\|_2 x^T
$$
$$
\nabla f(x) = \|x\|_2 x
$$
