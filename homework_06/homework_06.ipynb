Напишите функцию, которая принимает на вход параметры $N$ и $K$ и возвращает два числа: $x$ - cколько кусков эльфиского хлеба достанется каждому хоббиту, и $y$ - сколько кусков остаётся в корзинке.

```python
from PySide6.examples.widgets.itemviews.basicfiltermodel.basicsortfiltermodel import add_mail


def share_bread(N, K):
    # your code here
    if N < 1:
        return f"There are no hobbits"
    if N > K:
        return f"Not enough {K} bread for {N} hobbits: {N - K} pieces left"
    x = K // N  # floor division
    y = K % N  # module
    return x, y


# если в функции всё верно, то после выполнения этой строчки, не должно выскакивать ошибок
assert share_bread(N=0, K=0) == "There are no hobbits"
assert share_bread(N=1, K=0) == "Not enough 0 bread for 1 hobbits: 1 pieces left"
assert share_bread(N=3, K=14) == (4, 2)
```

## Задача 2

В копях Мории хоббиты нашли стену, на которой высечены разные натуральные числа. Согласно древним сказаниям, это даты сражений. Хоббиты знают, что сражения происходили только по високосным годам. Помогите хоббитам определить, является ли год с данным числом датой великого сражения. Если это так, то верните строку "YOU SHALL PASS", иначе верните "YOU SHALL NOT PASS". Напомним, что в соответствии с хоббитским календарем, год является високосным, если его номер кратен 4, но не кратен 100, а также если он кратен 400.

```python
def leap_year(year):
    # your code here
    if not isinstance(year, int) or year < 1:
        return f"Year must be positive integer"

    text_result = "YOU SHALL NOT PASS"
    if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
        text_result = "YOU SHALL PASS"

    return text_result


assert leap_year(25.4) == 'Year must be positive integer'
assert leap_year(-1) == 'Year must be positive integer'
assert leap_year(5) == 'YOU SHALL NOT PASS'
assert leap_year(8) == 'YOU SHALL PASS'
assert leap_year(200) == 'YOU SHALL NOT PASS'
assert leap_year(400) == 'YOU SHALL PASS'
assert leap_year(800) == 'YOU SHALL PASS'
```

Для могущественного магического ритуала Гендальфу необходимо быстро подсчитывать площадь своего амулета, который умеет менять размеры. Известно, что амулет имеет форму треугольника и Гендальф знает длину каждой из сторон. Напишите функцию, которая считает площадь амулета по трем сторонам.  

**Подсказка:** используйте формулу Герона
$$ S = \sqrt{p(p-a)(p-b)(p-c)} $$
$$ p = \frac{a + b + c}{2}$$


```python
def amulet_area(a, b, c):
    # your code here
    if a <= 0 or b <= 0 or c < 1 or a + b <= c or a + c <= b or b + c <= a:
        return f"Invalid triangle with sides a = {a}, b = {b}, c = {c}."

    p = (a + b + c) / 2
    S = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    return S


assert amulet_area(0, 4, 5) == "Invalid triangle with sides a = 0, b = 4, c = 5."
assert amulet_area(0, -33, 5) == "Invalid triangle with sides a = 0, b = -33, c = 5."
assert amulet_area(0, 4, 0) == "Invalid triangle with sides a = 0, b = 4, c = 0."
assert amulet_area(3, 4, 5) == 6
```

## Задача 4

Хоббиты собираются пешком идти до Мордора и им нужно подсчитать расстояние, которое им предстоит пройти. Хоббиты смогли вспомнить сразу несколько метрик расстояния: евклидово, манхэттена и косинусное, так что ваша задача - написать функцию под каждую из них. Важное условие - используйте только базовые функции для решения.


* Евклидово расстояние
$$ d(a, b) = \sqrt{\sum_i (a_i - b_i)^2} $$
* Расстояние Манхэттена
$$ d(a, b) = \sum_i |a_i - b_i| $$
* Косинусное расстояние
$$ d(a, b) = 1 - \frac{a^\top b}{||a||_2\cdot||b||_2}$$

```python
def cal_euclidean(a, b):
    ## Your code here
    if len(a) == 0 or len(b) == 0:
        return f"Invalid vectors with lengths {len(a)} and {len(b)}."
    distance = 0
    for i in range(len(a)):
        distance += (a[i] - b[i]) ** 2
    return distance ** 0.5


def cal_manhattan(a, b):
    ## Your code here
    if len(a) == 0 or len(b) == 0:
        return f"Invalid vectors with lengths {len(a)} and {len(b)}."
    distance = 0
    for i in range(len(a)):
        distance += abs(a[i] - b[i])
    return distance

 #realization with only basic functions
def cal_cosine(a, b):
    ## Your code here
    if len(a) == 0 or len(b) == 0:
        return f"Invalid vectors with lengths {len(a)} and {len(b)}."
    ab_mult = 0
    a_magnitude = 0
    b_magnitude = 0
    for i in range(len(a)):
        ab_mult += a[i] * b[i]
        a_magnitude += a[i] ** 2
        b_magnitude += b[i] ** 2
    distance_similarity = ab_mult / (a_magnitude ** 0.5 * b_magnitude ** 0.5)
    return 1 - distance_similarity
```

```python
import numpy as np

# debugging data
# a = [2,4]
# b = [4,2]

a = np.random.randint(-10, 10, size=10)
b = np.random.randint(-10, 10, size=10)
print(f"a: {a}")
print(f"b: {b}")
print(f"Euclidean distance: {cal_euclidean(a, b)}")
print(f"Manhattan distance: {cal_manhattan(a, b)}")
print(f"Cosine distance: {cal_cosine(a, b)}")

############################################
# Test Cosine distance with np.linalg.norm #
############################################
# Define the vectors
A = np.array(a)
B = np.array(b)
# Calculate cosine similarity
cos_similarity = np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))
# Calculate cosine distance
cos_distance = 1 - cos_similarity
print("Test Cosine distance with np.linalg.norm: {:.2f}".format(cos_distance))
```
